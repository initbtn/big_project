## 인덱스를 열로 변환 (reset_index)

DataFrame.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')

- 개요
reset_index메서드는 설정 인덱스를 제거하고 기본 인덱스(0,1,2, ... , n)으로 변경하는 메서드 입니다.

- 기본 사용법  

self.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')  
level : Multi Index의 경우 제거할 인덱스의 레벨을 설정할수 있습니다. 기본적으로 모든 레벨입니다.  
drop : 제거한 인덱스를 열에 추가할지 여부입니다. 기본값은 False로 제거된 인덱스는 열로 변환됩니다.  
inplace : pandas 공통 인수로, 원본을 변경할지 여부 입니다.  
col_level / col_fill : Multi Index의 경우 제거된 인덱스를 열에 추가할 때 레벨과 열의 이름을 설정할 수 있습니다.  
col_level을 통해 레벨을 설정하고, col_fill을통해 해당 열의 이름을 정합니다.  

```python
busan_data.reset_index(drop=True, inplace=True)
daegu_data.reset_index(drop=True, inplace=True)
daejeon_data.reset_index(drop=True, inplace=True)
gwangju_data.reset_index(drop=True, inplace=True)
seoul_data
```

## result_data에 결과 출력

```python
all_list = [seoul_data,busan_data,daegu_data,daejeon_data,gwangju_data]
result_data=pd.DataFrame(columns=seoul_data.columns)
for data in all_list:
    index=0
    loc=data['시장'].iloc[index]
    print(loc + "지역 데이터 확인~~~")
    for i in std_date:
        # print(data['Date'].iloc[index] , str(i))
        if(data['Date'].iloc[index]!=str(i)):
            # print(index)
            new_row = pd.DataFrame([[i, 0.0, 0, 0.0, loc]], columns = data.columns)
            # print(new_row)
            data=pd.concat([data.iloc[:index], new_row, data.iloc[index:]], ignore_index = True)
        index+=1
    # print(data)
    result_data=pd.concat([result_data,data],ignore_index=True)
    print(result_data)
    print(len(result_data))
```
