```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc,  font_manager
import matplotlib
fontname = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)
```

```python
get_data = pd.read_csv('data/TEMP/경락가격/양파/양파 월별 총거래(정제).csv', index_col=None)
get_weather = pd.read_csv('data/TEMP/날씨/기온 데이터.csv', index_col=None)
get_rain = pd.read_csv('data/TEMP/날씨/강수량 데이터.csv', index_col=None)
```

```python
get_weather['년월'] = get_weather['년월'].str.replace("\t\t\t", "")
```

```python
seoul_data = get_data[get_data['시장'] == '서울']
seoul_weather = get_weather[get_weather['지점'] == '서울']
seoul_rain = get_rain[get_rain['지점'] == '서울']

busan_data = get_data[get_data['시장'] == '부산']
busan_weather = get_weather[get_weather['지점'] == '부산']
busan_rain = get_rain[get_rain['지점'] == '부산']

daegu_data = get_data[get_data['시장'] == '대구']
daegu_weather = get_weather[get_weather['지점'] == '대구']
daegu_rain = get_rain[get_rain['지점'] == '대구']

daejeon_data = get_data[get_data['시장'] == '대전']
daejeon_weather = get_weather[get_weather['지점'] == '대전']
daejeon_rain = get_rain[get_rain['지점'] == '대전']

gwangju_data = get_data[get_data['시장'] == '광주']
gwangju_weather = get_weather[get_weather['지점'] == '광주']
gwangju_rain = get_rain[get_rain['지점'] == '광주']
```

```python
list = [seoul_data['총거래물량'], seoul_data['총거래금액'], seoul_data['평균거래금액'], seoul_weather['평균기온(℃)'], seoul_weather['평균최저기온(℃)'], seoul_weather['평균최고기온(℃)']]
column_names = ['총거래물량', '총거래금액', '평균가격', '평균기온', '평균최저기온', '평균최고기온']
df1 = pd.DataFrame(list).T
df1.columns = column_names
corr=df1.corr()
plt.title('월별 양파 서울과 날씨 상관관계')
corr_heatmap = sns.heatmap( corr, 
                           cbar=True, 
                           annot=True, 
                           annot_kws={'size': 20}, 
                           fmt='.2f', 
                           square=True, 
                           cmap='Blues',
                           vmin=0,
                           vmax=1
                           )
plt.savefig('data/월별 양파 서울과 날씨 상관관계.png')
```
